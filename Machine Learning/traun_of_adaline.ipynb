```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
y=df.iloc[:100,4].values
y=np.where(y==0,-1,1)
X=df.iloc[:100,[0,2]].values
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8, 4))

ada1=AdalineGD(n_iter=10,eta=0.1).fit(X,y)
ax[0].plot(range(1, len(ada1.cost_) + 1),(ada1.cost_), marker='o') 
ax[0].set_title('0.1')
ada2=AdalineGD(n_iter=10,eta=0.001).fit(X,y)
ax[1].plot(range(1,len(ada2.cost_)+1),(ada2.cost_), marker='o')
ax[1].set_title('0.001')
ada3=AdalineGD(n_iter=10,eta=0.0001).fit(X,y)
ax[2].plot(range(1,len(ada3.cost_)+1),np.log10(ada3.cost_), marker='o')
ax[2].set_title('0.0001')
```

```python
X_std = np.copy(X)
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std()
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
```

```python
ada1 = AdalineGD(n_iter=15, eta=0.01).fit(X_std, y)
plt.plot(range(1, len(ada1.cost_) + 1), ada1.cost_, marker='o') 
plt.title('0.01')
```
